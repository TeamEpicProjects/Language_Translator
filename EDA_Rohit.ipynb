```python
from nltk.tokenize import word_tokenize
words=[]
with open('story_pariksha.txt',encoding='utf8') as f:
    for line in f:
        words +=word_tokenize(line)
```

```python
from collections import Counter 
cnt = Counter(words)
cnt.most_common(10)
```

```python
from spacy.lang.hi import STOP_WORDS as STOP_WORDS_HI
non_stop_words = [word for word in words if word not in set(STOP_WORDS_HI) ]
non_stop_cnt = Counter(non_stop_words)
non_stop_cnt.most_common(10)
```

```python
fun=[',','.','|','!','।',';','`','~']
non_stop_words2=[a for a in non_stop_words if a[-1] not in fun]
non_stop_cnt2 = Counter(non_stop_words2)
non_stop_cnt2.most_common(10)
```

```python
text=' '.join(non_stop_words2)
text
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordcloud = WordCloud(
    width=300,
    height=300,
    max_font_size=100, 
    max_words=100,
    background_color="white", 
    regexp=r"[\u0900-\u097F]+", 
    font_path='mangal.ttf').generate(text)
plt.figure(figsize = (8, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
